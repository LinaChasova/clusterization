**DBSCAN** (Основанная на плотности пространственная кластеризация для приложений с шумами) группирует точки близкие друг другу основываясь на метрики расстояния и минимальным количеством точек в кластере. Таким образом, алгоритм может определять отдельные точки аутлаерами.

Каждая точка относится к след. классам:

- **Core**: такая точка, если в радиусе $\epsilon$ есть как минимум $minPoints$ точек;
- **Border**: такая точка, до которой может "дотянуться" *core* точка, но в радиусе $\epsilon$ есть меньше $minPoints$ точек;
- **Outlier**: не *core* и *border* точки.

**Алгоритм**

1. определить $\epsilon$ и $minPoints$;
2. начальная точка выбирается рандомно среди не посещаемых точек. если в радиусе $\epsilon$ не находится хотя бы $minPoints$ точек, начальная точка помечается *outlier*, повторяется шаг 2;
3. если в радиусе $\epsilon$ находится хотя бы $minPoints$ точек, начальная помечается как  *core*, происходит формирование кластера каждая соседняя точка присоединяется к кластеру, шаг 3 для каждой соседней точки;
4. повторить шаг 2-3 пока все точки не будут просмотрены.

```python
def dbscan(data: np.ndarray, eps: float, min_points: int = 5, metric: str = 'euclidean') -> np.ndarray:
    """ Проводит алгоритм кластеризации dbscan
    
    :param data: (n_samples, n_features) данные
    :param eps: параметр радиуса для нахождения соседей
    :param min_points: параметер минимальное кол-во точек в соседстве, чтобы образовать кластер
    :param metric: вид формулы расстояния
    :return: (n_samples) лейблы
    """
    def visit_point(index: int, first_cluster_point: bool = False):
        """ Посещаем точку
        
        :param index: индекс точки, которую рассматриваем
        :param first_cluster_point: яв-ся ли точка первой точкой потенциального кластера
        """
        if is_visited[index]:  # точка либо core либо border
            return

        # находим соседей в радиусе eps
        distances = cdist(data[index].reshape(1, 2), data, metric)
        neighbour_indices = np.where(
            np.any((distances <= eps) & (distances > 0), axis=0)
        )[0]

        if len(neighbour_indices) >= min_points:  # core
            is_visited[index] = True   # точка core
            labels[index] = cluster_id  # помечаем ее кластер

            for neighbour_index in neighbour_indices:  # проходим по точкам соседей
                visit_point(neighbour_index)
        # border, только если не первая точка кластера
        elif len(neighbour_indices) < min_points and len(neighbour_indices) != 0 and not first_cluster_point:
            is_visited[index] = True  # точка border
            labels[index] = cluster_id # помечаем ее кластер
        else: # outlier
            labels[index] = outlier_id

        return

    cluster_id, outlier_id = -1, -1

    # помечаем если точки были посещены (помимо аутлаеров)
    is_visited = np.full((data.shape[0]), False)
    labels = np.full((data.shape[0]), np.inf)    # лейблы каждой точки

    for index in range(len(data)):  # идем по каждой точке
        cluster_id += 1
        visit_point(index, True)  # определяем ее к кластеру или аутлаерам
        
    clusters = np.unique(labels)
    for n_cluster, cluster_id in enumerate(clusters):
        labels[labels == cluster_id] = n_cluster - 1

    return labels
```

```python
RANDOM_STATE = 0

centers = [[1, 1], [-1, -1], [1, -1]]
centers = np.reshape(centers, (3, 2))
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=RANDOM_STATE
)
X = StandardScaler().fit_transform(X)

plt.scatter(X[:, 0], X[:, 1], s=50, c="b")
plt.show()
```

$\epsilon$ один из параметров, который нужно подбирать, чтобы алгоритм сработал верно. Если значение выбрано слишком маленьким бОльшая часть точек не будет окластеризирована и будет отмечена как аутлаеры. Если значение выбрано слишком большим то многие кластеры будут объединяться в один.

$\epsilon$ должна подбираться в зависимости от расстояния всех данных (используя, например, $k$-distance граф), однако в основном своем более маленькие значения $\epsilon$ предпочтительно.

При использовании $k$-distance граф, строя расстояния до $k = minPoints - 1$ соседа отсортированным с самого большого до маленького; значение $\epsilon$ можно вытащить с помощью алгоритма локтя.

```python
def plot_resulted_clusters(
    data: np.ndarray,
    labels: np.ndarray,
    n_row_plots: int,
    n_column_plots: int,
    n_plot: int,
    eps: float = .5,
    min_points: int = 5,
    metric: str = 'euclidean'
):
    """ Отрисовка данных с разными параметрами

    :param data: (n_samples, n_features) данные
    :param labels: (n_samples) лейблы
    :param n_row_plots: количество графиков в строке
    :param n_column_plots: количество графиков в колонке
    :param n_plot: номер графика
    :param eps: параметр радиуса для нахождения соседей
    :param min_points: параметер минимальное кол-во точек в соседстве, чтобы образовать кластер
    :param metric: вид формулы расстояния
    """   
    plt.subplot(n_row_plots, n_column_plots, n_plot)
    scatter = plt.scatter(data[:, 0], data[:, 1], c=labels, s=50, cmap='viridis')
    plt.legend(handles=scatter.legend_elements()[0], labels=scatter.legend_elements()[1])
    plt.title(f'Результат кластеризации | {eps:.2f} | {min_points}\n {metric}')
```

```python
params = [0.1, 0.15, 0.2, 0.25, 0.3]

n_column_plots = 3
n_row_plots = len(params) / n_column_plots if len(
    params) % n_column_plots == 0 else len(params) // n_column_plots + 1
plt.figure(figsize=(n_column_plots * 5, n_row_plots * 6))

for n_row, eps in enumerate(params):
    n_plot = n_row // n_column_plots * n_column_plots + (n_row % n_column_plots) + 1
    labels = dbscan(X, eps=eps, min_points=9)
    plot_resulted_clusters(X, labels, n_row_plots=n_row_plots,
                           n_column_plots=n_column_plots, n_plot=n_plot, eps=eps, min_points=9)
```

```python
neighbors = NearestNeighbors(n_neighbors=2).fit(X)
distances, indices = neighbors.kneighbors(X)

distances = np.sort(distances[:, -1].flatten())[::-1]
plt.plot(distances)

kl = KneeLocator(range(0, len(distances)), distances, curve='convex', direction='decreasing')
round(kl.elbow_y, 2)
```

Минимальное значение $minPoints$ можно высчитать из размерности данных $D$, $minPoints \geq D + 1$. Большие значения лучше использовать для более шумных данных.

Значение $minPoints = 1$ не имеет смысла, т.к. таким образом мы получим каждую точку как отдельный кластер. Когда $minPoints \leq 2$ получается результат Agglomerative clustering с использованием связности single linkage, и дендограма обрезана на высоте $\epsilon$. Так $minPoints$ должно быть хотя бы 3.

Можно использовать практическое правило, где $minPoints = 2*D$, но иногда нужно использовать большие значения для больших данных, для шумных данных или данных с большим количеством повторений.

```python
params = [3, 4, 5, 6, 7, 8, 9]
eps = 0.16

n_column_plots = 3
n_row_plots = len(params) / n_column_plots if len(
    params) % n_column_plots == 0 else len(params) // n_column_plots + 1

plt.figure(figsize=(n_column_plots * 5, n_row_plots * 6))

for n_row, min_points in enumerate(params):
    n_plot = n_row // n_column_plots * \
        n_column_plots + (n_row % n_column_plots) + 1
    labels = dbscan(X, eps=eps, min_points=min_points)
    plot_resulted_clusters(X, labels, n_row_plots=n_row_plots,
                           n_column_plots=n_column_plots, n_plot=n_plot, eps=eps, min_points=min_points)
```

Так же можно использовать разные метрики для подсчета **расстояния**.

- Euclidean: $$d(x, y) = \sqrt{\sum_{i=1}^n{(y_i - x_i)^2}}$$
- Canberra: $$d(x, y) = \sum_{i=1}^n \frac{|x_i - y_i|}{|x_i| + |y_i|}$$
- Chebyshev: $$d(x, y) = \max_i |x_i - y_i|$$

```python
params = ['euclidean', 'canberra', 'chebyshev']
eps = .2
min_points = 4

n_column_plots = 3
n_row_plots = len(params) / n_column_plots if len(
    params) % n_column_plots == 0 else len(params) // n_column_plots + 1
plt.figure(figsize=(n_column_plots * 5, n_row_plots * 6))

for n_row, metric in enumerate(params):
    n_plot = n_row // n_column_plots * \
        n_column_plots + (n_row % n_column_plots) + 1
    labels = dbscan(X, eps=eps, min_points=9, metric=metric)
    plot_resulted_clusters(X, labels, n_row_plots=n_row_plots,
                           n_column_plots=n_column_plots, n_plot=n_plot,
                           eps=eps, min_points=min_points, metric=metric)
```

Существует имплементация в библиотеке [scikit-learn](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html).

Примите во внимание, что в библиотечном алгоритме параметр `min_samples` для определения минимальной плотности кластера так же рассматривает core точку.

```python
eps, min_samples = .3, 10

db = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
labels = db.labels_

plt.figure(figsize=(5, 6))
plot_resulted_clusters(X, labels, n_row_plots=1, n_column_plots=1, n_plot=1,
                       eps=eps, min_points=min_samples, metric=metric)
```

**Итог**

| Параметры | Плюсы | Минусы |
| :-------- | :---- | :----- |
| - `eps`: как близко точки должны находится, чтобы принодлежать к одному кластеру;<br> - `min_samples`: минимальное количество точек, чтобы определить эту группу к кластеру;<br> - `metric`: формула расстояния | - не чувствителен к аутлаерам;<br> - можно использовать на данных с кластерами разных размеров;<br> - не надо укащывать количество кластеров. | - иногда подбор $\epsilon$ требует знание области;<br> - плохо работает с кластерами разной плотности.
