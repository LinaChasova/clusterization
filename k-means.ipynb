K-means является одним из самых популярных алгоритмов кластеризации. Основная идея разделить данные на $k$ кластеров, определяя каждую точку к ближайшему центроиду кластера. Для достижения лучшего результата алгоритм чередуется между определением данных к ближайшим центроидам и выбиранием центроидов.

**Имплементация алгоритма**

1. Инициализируем центроиды;
2. Считаем расстояние от каждой точки до каждого центроида;
3. Определяем точку к кластеру ближайшего центроида;
4. Выбираем новые центроиды как середина каждого кластера;
5. Повторяем шаги 2-4 пока не достигли максимальной итерации или разница между центроидами равна нулю.

```python
def kmeans(
    data: np.ndarray,
    n_clusters: int = 5,
    init: str = 'kmeans++',
    max_iter: int = 300,
    distance: str = 'euclidean',
    random_state: int = None,
    epsilon: float = 1e-4
) -> tuple[list, list]:
    """Провести кластеризацию с помощью алгоритма k-means

    :param data: (n_samples, n_features) данные
    :param n_clusters: количество кластеров и центроидов
    :param init: вид инициализации центроидов кластеров
    :param max_iter: максимальное количество итераций
    :param distance: алгоритм используемый для подсчета расстояния между точками
    :param random_state: сид
    :param epsilon: допустимое значения для разницы между значениями 
                    центроидов используется для сходимости алгоритма
    :return: centroids_history (n_iters, n_clusters, n_features) историю центроидов для отрисовки,
             labels (n_samples, ) лейблы точек
    """
    # инициализируем центориды кластеров
    if init == 'kmeans++':
        centroids = get_kmeanspp_centroids(
            data, n_clusters, distance, random_state)
    elif init == 'random':
        centroids = get_random_centroids(data, n_clusters, random_state)
    elif init == 'random_data':
        centroids = get_random_data_centroids(data, n_clusters, random_state)
    else:
        raise ValueError(f'no init algorithm for {init}')

    # сохраняем историю центроидов для отрисовки
    centroids_history = []
    centroids_history.append(centroids)
    for i in range(max_iter):
        # считаем расстояние между данными и центроидами
        distances = cdist(X, centroids, metric=distance)
        # определяем ближайший центроид
        new_labels = distances.argmin(axis=1)

        # выбираем новые центроиды в центре полученных кластеров
        # centroids = centroids.copy()
        centroids = np.array([data[new_labels == i].mean(0)
                              for i in range(n_clusters)])
        
        # считаем норму разницы двух вариантов центроидов
        diff = centroids - centroids_history[-1]
        diff = np.linalg.norm(diff, ord='fro')

        if abs(diff.sum()) <= epsilon:
            break

        centroids_history.append(centroids)
        labels = new_labels

    return centroids_history, labels
```

Рандомно инициализируем датасет со значениями кластера

```python
RANDOM_STATE = 8

centers = [[1, 1], [-1, -1], [0.5, -0.5]]
centers = np.reshape(centers, (3, 2))

X, true_labels = make_blobs(
    n_samples=3000,
    centers=centers,
    cluster_std=0.7,
    random_state=RANDOM_STATE
)

plt.scatter(X[:, 0], X[:, 1], s=5, c="b")
plt.show()
```

**Функции для инициализации центроидов**

Один из недостатков алгоритма является задача начальной инициализации центроидов. Самые простые алгоритмы являются либо случайным семплингов точек, либо выбор случайных точек из датасета. Таким образом, если центроид инициализирован как «удаленная» точка, он может просто остаться без связанных с ним точек, и в то же время более одного кластера могут оказаться связанными с одним центроидом. Точно так же несколько центроидов могут быть инициализированы в одном кластере, что приведет к плохой кластеризации.

Для решения этой проблемы можно использовать алгоритм k-means++.
1. Рандомно выбрать первый центроид среди $x_1, \ldots, x_n$;
2. Для каждой $x_1, \ldots, x_n$ посчитать расстояние до ближайшего выбранного центроида;
3. Выбрать следующий центроид среди $x_1, \ldots, x_n$ таким образом, что вероятность выбора определенной точки прямо порпорциональна расстоянию от ближайшего выбранного центроида;
4. Повторить шаги 2 и 3 пока $k$ центроидов не выбраны.

```python
def get_kmeanspp_centroids(data: np.ndarray, n_clusters: int, distance: str,
                           random_state: int) -> np.ndarray:
    """K-means++ алгоритм для инициализации центроидов

    :param data: (n_samples, n_features) данные
    :param n_clusters: количество кластеров и центроидов
    :param distance: алгоритм используемый для подсчета расстояния между точками
    :param random_state: сид
    :return: centroids (n_clusters, n_features) центроиды
    """
    np.random.seed(seed=random_state)
    n_samples, n_features = data.shape[0], data.shape[1]
    
    # выбираем случайно первый центроид
    index = np.random.choice(n_samples, 1)
    centroids = data[index].reshape((1, 2))
    
    i = 1
    while i != n_clusters:
        # считаем расстояние до последне выбранного центроида и нормализуем расстояние
        distances = cdist(data, centroids[-1, :].reshape((1, 2)), metric=distance)
        distances = distances / distances.sum()
        
        # выбираем следующий центроид с вероятностью зависимой от расстояния до прошлего центроида
        index = np.random.choice(n_samples, 1, p=distances.reshape(len(distances),))
        centroids = np.append(centroids, data[index], axis=0)     
        i += 1
    
    return centroids
```

```python
def get_random_centroids(
    data: np.ndarray, n_clusters: int, random_state: int
) -> np.ndarray:
    """Инициализируем центроиды рандомно семплируя точки

    :param data: (n_samples, n_features) данные
    :param n_clusters: количество кластеров и центроидов
    :param random_state: сид
    :return: centroids (n_clusters, n_features) центроиды
    """
    np.random.seed(seed=random_state)

    n_features = data.shape[1]
    centroids = np.random.normal(size=n_clusters * n_features)
    centroids = centroids.reshape((n_clusters, n_features))

    return centroids
```

```python
def get_random_data_centroids(data: np.ndarray, n_clusters: int, random_state: int) -> np.ndarray:
    """Инициализируем центроиды рандомно выбирая точки среди данных

    :param data: (n_samples, n_features) данные
    :param n_clusters: количество кластеров и центроидов
    :param random_state: сид
    :return: centroids (n_clusters, n_features) центроиды
    """
    np.random.seed(seed=random_state)

    n_samples = data.shape[0]
    index = np.random.choice(n_samples, n_clusters, replace=False)
    centroids = data[index]

    return centroids
```

Рассмотрим различные виды инициализации. Графики ниже показывают центроиды и результат кластеризации сразу после инициализации и в конце алгоритма.

```python
def plot_diff_init(init: str, i: int):
    """ Отрисовка инициализации центроидов и финальный вариант центроидов
    
    :param init: {'kmeans++', 'random', 'random_data'} имя алгоритма инициализации
    :param i: номер графика
    """
    centroids_history, labels = kmeans(
        X, n_clusters=3, init=init, random_state=RANDOM_STATE
    )

    # начальные лейблы
    distances = cdist(X, centroids_history[0])
    init_labels = distances.argmin(axis=1)

    plt.subplot(4, 2, i)
    plt.scatter(X[:, 0], X[:, 1], c=init_labels, s=5, cmap="viridis")
    plt.plot(centroids_history[0][:, 0], centroids_history[0][:, 1], "rX")
    plt.title(f"Первая итерация с инициализацией {init}")

    # результат лейблы
    plt.subplot(4, 2, i + 1)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=5, cmap="viridis")
    plt.plot(centroids_history[-1][:, 0], centroids_history[-1][:, 1], "rX")
    plt.title(f"Результат с инициализацией {init}")
```

```python
plt.figure(figsize=(10, 15))

# -- kmeans++ инициализация
plot_diff_init('kmeans++', 1)
# -- random инициализация
plot_diff_init('random', 3)
# -- random_data инициализация
plot_diff_init('random_data', 5)

# -- результат лейблы
plt.subplot(4, 2, 7)
plt.scatter(X[:, 0], X[:, 1], c=true_labels, s=5, cmap='viridis')
plt.plot(centers[:, 0], centers[:, 1], "rX")
plt.title('Настоящая классификация')

plt.show()
```

Так же можно использовать разные метрики для подсчета **расстояния**.

- Euclidean: $$d(x, y) = \sqrt{\sum_{i=1}^n{(y_i - x_i)^2}}$$
- Cosine: $$d(x, y) = 1 - \frac{x y}{\|x\| \|y\|}$$
- Canberra: $$d(x, y) = \sum_{i=1}^n \frac{|x_i - y_i|}{|x_i| + |y_i|}$$
- Chebyshev: $$d(x, y) = \max_i |x_i - y_i|$$

```python
def plot_diff_distance(distance: str, i: int):
    """ Отрисовка результатов кластеризации в зависимости от функции расстояния

    :param distance: имя алгоритма расстояния
    :param i: номер графика
    """
    centroids_history, labels = kmeans(
        X, n_clusters=3, distance=distance, random_state=RANDOM_STATE
    )

    plt.subplot(3, 2, i)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=5, cmap="viridis")
    plt.plot(centroids_history[-1][:, 0], centroids_history[-1][:, 1], "rX")
    plt.title(f"Результат с расстоянием {distance}")
```

```python
plt.figure(figsize=(10, 12))

# -- euclidean
plot_diff_distance('euclidean', 1)
# -- cosine
plot_diff_distance('cosine', 2)
# -- canberra
plot_diff_distance('canberra', 3)
# -- chebyshev
plot_diff_distance('chebyshev', 4)

# -- результат лейблы
plt.subplot(3, 2, 5)
plt.scatter(X[:, 0], X[:, 1], c=true_labels, s=5, cmap='viridis')
plt.plot(centers[:, 0], centers[:, 1], "rX")
plt.title('Настоящая классификация')

plt.show()
```

**Выбор количества кластеров**

Основная задача алгоритма оптимизировать **инерцию** или внутре-кластерную сумму квадратов: $$J(C) = \sum_{k=1}^K \sum_{i\in C_k} \| x_i - \mu_k \| \rightarrow \min_C$$ где $C$ это множество кластеров длиной `n_clusters`, $\mu_k$ это центроид кластера $C_k$ 

Однако, в таком случае $J(C)$ является минимум, когда `n_clusters == n_samples`; поэтому оптимальное количество кластеров можно выбрать так, после которого значение $J(C)$ уменьшается не так резко. Это называется алгоритм локтя.

В следующих примерах, будет использоваться функция [k-means библиотеки scikit-learn](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html).

```python
inertia = []
for k in range(1, 8):
    k_means = KMeans(n_clusters=k, random_state=RANDOM_STATE).fit(X)
    inertia.append(np.sqrt(k_means.inertia_))

plt.plot(range(1, 8), inertia, marker='s')
plt.xlabel('$k$')
plt.ylabel('$J(C)$');
```

По графику видно, что оптимальное количество кластеров равно трем. После этого график начинает уменьшаться не так резко.

Определить точку изгиба на кривой $J(C)$ не всегда просто. Можно использовать библиотеку [kneed](https://github.com/arvkevi/kneed) для программной идентификации точки изгиба:

```python
kl = KneeLocator(range(1, 8), inertia, curve='convex', direction='decreasing')
kl.elbow
```

**Mini Batch K-Means**

Mini Batch K-Means - это вариант алгоритма K-Means. Он обрабатывает данные частями для сокращения времени вычислений. Части данных собираются случайным образом на каждой итерации алгоритма. Как правило, Mini Batch K-Means дают результаты лишь немного хуже, чем стандартный алгоритм.

1. Выбирается выборка из $b$ точек и они присваиваются к ближайшим центроидам;
2. Обновляются центроиды. В отличие от K-Means, обновление происходит для каждой точки в выборке - для каждой точки его присвоенный центроид обновляется путем подтсчета среднего потокового значения этой точки и всех прошлых точек присвоенных к этом центроиду. Таким образом уменьшает флуктуация центроида с течением времени;
3. Шаги 1-2 повторяются пока не достигнута сходимость или максимальное количество итераций.

Mini Batch K-Means сходится быстрее, чем K-Means, но качество результатов снижается. На практике эта разница в качестве может быть довольно небольшой.

```python
k_means = KMeans(n_clusters=3, n_init=10)
t0 = time.time()
k_means.fit(X)
t_batch = time.time() - t0

mbk = MiniBatchKMeans(
    n_clusters=3, batch_size=45, n_init=10, max_no_improvement=10, verbose=0,
)
t0 = time.time()
mbk.fit(X)
t_mini_batch = time.time() - t0

kmeans_cluster_centers = k_means.cluster_centers_
distances = cdist(k_means.cluster_centers_, mbk.cluster_centers_)
order = distances.argmin(axis=1)
mbk_cluster_centers = mbk.cluster_centers_[order]

distances = cdist(X, kmeans_cluster_centers)
kmeans_labels = distances.argmin(axis=1)

distances = cdist(X, mbk_cluster_centers)
mbk_labels = distances.argmin(axis=1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=5, cmap="viridis")
plt.plot(kmeans_cluster_centers[:, 0], kmeans_cluster_centers[:, 1], "rX")
plt.title(
    f"Результат K-Means\ntrain time: {t_batch:.2f}s\ninertia: {k_means.inertia_:.2f}"
)

plt.subplot(1, 3, 2)
plt.scatter(X[:, 0], X[:, 1], c=mbk_labels, s=5, cmap="viridis")
plt.plot(mbk_cluster_centers[:, 0], mbk_cluster_centers[:, 1], "rX")
plt.title(
    f"Результат Mini Batch K-Means\ntrain time: {t_mini_batch:.2f}s\ninertia: {mbk.inertia_:.2f}"
)

different = mbk_labels == 4
for k in range(3):
    different += (kmeans_labels == k) != (mbk_labels == k)

plt.subplot(1, 3, 3)
identic = np.logical_not(different)
plt.scatter(X[:, 0], X[:, 1], c=different, s=5, cmap="viridis")
plt.title(f"Разница результатов")

plt.show()
```

**Итог:**

| Параметры | Минусы | Плюсы |
| :--------- | :------ | :----- |
| - `n_clusters`: число кластеров; | - легко имплементировать;<br> - можно использовать на больших данных с помощью Mini-Batch K-Means;<br> - гарантированная сходимость. | - выбор кластеров `n_clusters`;<br> - зависимость от начальной инициализации центроидов;<br> - проблемы с кластеризацией данных, где кластеры разного размера и плотности;<br> - аутлаеры могут сильно влиять на кластеры;<br> - с увеличением размерности данных, страдает от curse of dimensionality. |
